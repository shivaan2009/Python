# If-Elif-Else Statement


```python
if 25 > 10:
    print('It works')
```

```python
if (25 < 10) or (1 < 3):
    print('It works')
elif 25 > 30:
    print('Elif worked')
elif 25 < 21:
    print('Elif 2 worked')
elif 25 < 40:
    print('Elif 3 worked')
else:
    print('It did not work')
```

```python
print('It works') if 10>30 else print('It does not work')
```

```python
if (25 < 10) or (1 < 3):
    print('It works')
    if 10 > 5:
        print('nested if statement works')
elif 25 > 30:
    print('Elif worked')
elif 25 < 21:
    print('Elif 2 worked')
elif 25 < 40:
    print('Elif 3 worked')
else:
    print('It did not work')
```
