# Merge, Join, and Concatenate


```python
import pandas as pd
```

```python
df1 = pd.read_csv(r"C:\Users\shiva\jupyter\Alex Analyst Python\Pandas\csv file\LOTR.csv")
df1
```

```python
df2 = pd.read_csv(r"C:\Users\shiva\jupyter\Alex Analyst Python\Pandas\csv file\LOTR 2.csv")
df2
```

```python
df1.merge(df2) # does inner join(default)
```

```python
df1.merge(df2, how = 'inner', on = ['FellowshipID','FirstName'])
```

```python
df1.merge(df2, how = 'outer')
```

```python
df1.merge(df2, how = 'left')
```

```python
df1.merge(df2, how = 'right')
```

```python
df1.merge(df2, how = 'cross')
```

```python
df1.join(df2, on = 'FellowshipID',how = 'outer', lsuffix = '_left',rsuffix = '_right')
```

```python
df4 = df1.set_index('FellowshipID').join(df2.set_index('FellowshipID'), lsuffix = '_left',rsuffix = '_right',how = 'outer')
df4
```

```python
pd.concat([df1,df2], join = 'outer',axis = 1) # sits on top of each other
```

```python
df1.append(df2)
```
